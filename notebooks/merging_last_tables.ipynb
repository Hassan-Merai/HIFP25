```python
import s3fs
fs = s3fs.S3FileSystem()

# Get list of all CSV files
all_files = fs.glob('s3://medicare-fraud-data-25-05-2025/merged_ready/train/*.csv')

# Read and concatenate
dfs = [pd.read_csv(f's3://{f}') for f in all_files]
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
merged_df
```

```python
# Get the top feature columns (e.g., Top1_Feature, Top2_Feature, etc.)
top_feature_columns = [col for col in shap_df.columns if "Top" in col and "Feature" in col]

# Create a list of all unique top features
all_top_features = []
for col in top_feature_columns:
    all_top_features.extend(shap_df[col].unique())
all_top_features = list(set(all_top_features))  # Remove duplicates
```

```python
all_top_features
```

```python
# Select only the relevant columns from merged_df (Provider + top features)
#merged_subset = merged_df[["Provider"] + all_top_features]

# Merge with shap_df
merged_result = pd.merge(
    shap_df,
    merged_df,
    on="Provider",
    how="left"
)
```

```python
for i, feature_col in enumerate(top_feature_columns, start=1):
    # Get the feature name for each row
    feature_names = shap_df[feature_col]
    
    # Look up the actual value from merged_df
    actual_values = [
        merged_df.loc[merged_df["Provider"] == provider, feature].values[0]
        for provider, feature in zip(shap_df["Provider"], feature_names)
    ]
    
    # Add to the DataFrame
    merged_result[f"Top{i}_ActualValue"] = actual_values
```
